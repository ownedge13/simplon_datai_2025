5 Tracer une courbe sinuso√Ødale en utilisant NumPy.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)


fig, ax = plt.subplots()

ax.plot(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
