3. Tracer un diagramme à barres à partir des données suivantes : 
categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

```python
import matplotlib.pyplot as plt
import numpy as np

species = ("A", "B", "C", "D", "E")

# make data:
#x = 0.5 + np.arange(5)
x = 0.5 + np.arange(len(species))
y = [25, 30, 15, 20, 10]

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.5)

ax.set(xlim=(0, 5), xticks=np.arange(1, 5),
       ylim=(0, 30), yticks=np.arange(1, 35))

#ax.set_xticks(x + species)

plt.show()
```
