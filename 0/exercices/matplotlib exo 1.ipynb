1. Tracer un graphe linéaire des points (x, y) où x = [0, 1, 2, 3, 4] et y = [1, 4, 3, 8, 5].

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [1, 4, 3, 8, 5]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

2. Tracer un histogramme à partir de données aléatoires générées par NumPy.

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
np.random.seed(5)
x = 5 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x)

plt.show()
```

3. Tracer un diagramme à barres à partir des données suivantes : 
categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

```python
import matplotlib.pyplot as plt
import numpy as np

# make data:
x = 0.5 + np.arange(5)
y : [25, 30, 15, 20, 10]

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.5)

ax.set(xlim=(0, 5), xticks=np.arange(1, 5),
       ylim=(0, 5), yticks=np.arange(1, 5))

plt.show()
```
