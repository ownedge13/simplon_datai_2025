6 Tracer deux courbes sur le même graphe à partir des données suivantes :
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y2)
ax.plot(x, y1)

ax.set(xticks=np.arange(0, 11), yticks=np.arange(0, 3))

plt.show()
```
