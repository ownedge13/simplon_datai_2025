```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=1000, n_features=1, noise=10)
```

```python
# plot
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# size and color:
sizes = np.random.uniform(15, 80, len(X))
colors = np.random.uniform(15, 80, len(X))

plt.scatter(X, y, s=sizes, c=colors, vmin=0, vmax=100)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33)
```

```python
plt.scatter(X_train, y_train, label = 'train')
plt.scatter(X_test, y_test, label = 'test')
plt.legend()
```

R2

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)

# plt.scatter(X_train, y_train, color ='b')
plt.scatter(X_train, y_train, label = 'train')

# plt.plot(X_test, y_pred, color ='k')
plt.plot(X_test, y_pred, label = 'test', color ='k')
plt.legend()
plt.show()
```

MSE

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
#autre methode
# np.mean((y_test - y_pred)**2)
```

RMSE

```python
mean_squared_error(y_test, y_pred, squared=False)
```

```python
y2 = y ** 2
```

```python
from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X, y2, test_size=0.33)

regr = LinearRegression()
regr.fit(X_train2, y_train2)
print(regr.score(X_test2, y_test2))
```

```python
y_pred2 = regr.predict(X_test2)

plt.scatter(X_test2, y_test2, color ='b')
plt.scatter(X_train2, y_train2, color ='r')

plt.plot(X_test2, y_pred2, color ='k')

plt.show()
```
