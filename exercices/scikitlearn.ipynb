```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

creation data de 1000 points

```python
X, y = make_regression(n_samples=1000, n_features=1, noise=10)
```

affichage des 1000

```python
sizes = np.random.uniform(15, 80, len(X))
colors = np.random.uniform(15, 80, len(X))

plt.scatter(X, y, s=sizes, c=colors, vmin=0, vmax=100)
```

creation d'un jeu de train et un jeu de test 33/66

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33)
```

affichage train et test

```python
plt.scatter(X_train, y_train, label = 'train')
plt.scatter(X_test, y_test, label = 'test')
plt.legend()
```

```python
X_train
```

R2

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)

# plt.scatter(X_train, y_train, color ='b')
plt.scatter(X_train, y_train, label = 'train')

# plt.plot(X_test, y_pred, color ='k')
plt.plot(X_test, y_pred, label = 'prediction', color ='k')
plt.legend()
plt.show()
```

MSE

```python
mean_squared_error(y_test, y_pred)
#autre methode
# np.mean((y_test - y_pred)**2)
```

RMSE

```python
mean_squared_error(y_test, y_pred, squared=False)
```

passage de y au carré

```python
y2 = y ** 2
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X, y2, test_size=0.33)

regr = LinearRegression()
regr.fit(X_train2, y_train2)
print(regr.score(X_test2, y_test2))
```

```python
y_pred2 = regr.predict(X_test2)

plt.scatter(X_test2, y_test2, label = 'test', color ='b')
plt.scatter(X_train2, y_train2, label = 'train', color ='r')

plt.plot(X_test2, y_pred2, label = 'prediction', color ='k')
plt.legend()
plt.show()
```

```python
y3 = y ** 2
```

polynomial sur X

```python
X2_train, X2_test, y3_train, y3_test = train_test_split(
     X, y3, test_size=0.33)
poly = PolynomialFeatures(degree=2)
X_pol = poly.fit_transform(X2_train)

regr = LinearRegression()
regr.fit(X_pol, y3_train)
print(regr.score(X_pol, y3_train))
```

affichage prediction

```python
y3_pred = regr.predict(X_pol)

plt.scatter(X2_test, y3_test, label = 'test', color ='b')
plt.scatter(X2_train, y3_train, label = 'train', color ='r')

plt.scatter(X2_train, y3_pred, label = 'prediction', color ='k')
# sorted_indices = np.argsort(x_test.T[0])
# plt.plot(x_test[sorted_indices], y_hat[sorted_indices], c="red")
plt.legend()
plt.show()
```
