8 Tracer une courbe polynomiale à partir des données suivantes :
x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
data = 3 * x**3 + 2 * x**2 - 8 * x + 5

fig, ax = plt.subplots()

ax.plot(x, data)
```
