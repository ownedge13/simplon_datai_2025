formatage table Sub_Category avec ID primaire et ID2 étrangère (Product_Category)

```python
#raz previous formatage
salespd = pd.read_csv(salescsv)
salesdropmonth = salespd.drop('Month', axis=1)
salesdropday = salesdropmonth.drop('Day', axis=1)
salesdropyear = salesdropday.drop('Year', axis=1)
newcolonnes = ['Date_', 'Customer_Age', 'Age_Group', 'Customer_Gender', 'Country', 'State', 'Product_Category', 'Sub_Category', 'Product', 'Order_Quantity', 'Unit_Cost', 'Unit_Price', 'Profit', 'Cost', 'Revenue']
salesdropyear.columns = newcolonnes

#formatage table Sub_Category avec ID primaire et ID2 étrangère (Product_Category)

salesdropyear['Product_Category'] = salesdropyear['Product_Category'].factorize()[0] + 1
variables = ['Product_Category', 'Sub_Category']
subsetcat = salesdropyear[variables]
subsetcat2 = subsetcat.drop_duplicates(subset=['Sub_Category'])
new_subsetcat=subsetcat2.assign(ID2=1)
new_subsetcat['ID2'] = np.arange(len(new_subsetcat))+1
newcolonnes = ['ID2', 'Sub_Category', 'ID']
new_subsetcat.columns = newcolonnes
new_subsetcat
```
